```python
import pandas as pd 
df = pd.read_excel('https://api.worldbank.org/v2/en/indicator/EN.ATM.CO2E.KT?downloadformat=excel', skiprows=3)

print(df)
```

```python
#Drop unnecessary columns
df1 = df.drop(df.loc[:, 'Country Code':'1989'].columns,axis = 1)
df1 = df1.drop(df1.loc[:, '2020':'2021'].columns,axis = 1)
#Drop NaN Values
df1 = df1.dropna()
```

```python
#selecting data for two countries: 
#country1: United States
country1_df = df1[df1['Country Name'] == 'United States']
#country2: China
country2_df = df1[df1['Country Name'] == 'China']

```

```python
R = country1_df.T
```

```python
headers = R.iloc[0]
#Create DataFrame
R = pd.DataFrame(R.values[1:], columns=headers)
```

```python
R
```

```python
#Insert year column
R.insert(0, "Year", ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'], True)

R
```

```python
#Convert to numeric
df_fitting = R[['Year', 'United States']].apply(pd.to_numeric, 
                                               errors='coerce')
```

```python
#Select only numerical values
m = df_fitting.dropna().values
```

```python
#Fit curve for country1
import numpy as np
from scipy.optimize import curve_fit

x1 = m[:,0] #Selecting years from 1990 - 2019
y1 = m[:,1] #Selecting CO2 emissions
```

```python
import numpy as np

from scipy.optimize import curve_fit

import matplotlib.pyplot as plt


# Define logistic function

def logistic_func(x, k, x0, L):

    return L / (1 + np.exp(-k*(x-x0)))
```

```python
#Fit curve
popt, pcov = curve_fit(logistic_func, x1, y1, p0=[2, 2000, 5775810])
```

```python
print(popt)
```

```python
print(pcov)
```

```python
# Calculate the predicted values for 10 and 20 years

x10 = 2000 + 10

x20 = 2000 + 20

y10 = logistic_func(x10, *popt)

y20 = logistic_func(x20, *popt)
```

```python
# Create a plot
x = np.linspace(1990, 2020, 30)
y = logistic_func(x, *popt)
```

```python
plt.plot(x, y, 'r-', label='Fitting curve')
plt.scatter(x1, y1, label='Data')

plt.scatter(x10, y10, label='Data 10', color='orange')

plt.scatter(x20, y20, label='Data 19', color='green')

plt.xlabel('Year')

plt.ylabel('CO2 emission (kt)')

plt.legend()

plt.show()
```

```python
#prediction of CO2 emission by 2030
x30 = 2000 + 30
y30 = logistic_func(x30, *popt)
print("The CO2 emission of the United States by 2030 is predicted to be", y30, "kt")
```

```python
# Estimate lower and upper limits of the confidence range
def err_ranges(x, y, popt):
    def logistic_func(x, k, x0, L):

        return L / (1 + np.exp(-k*(x-x0)))
    n = len(x)
    y_pred = logistic_func(x, *popt)
    err = np.sum((y - y_pred)**2)
    var = err/(n-len(popt))
    std = np.sqrt(var)
    up = logistic_func(x, *(popt + std))
    low = logistic_func(x, *(popt - std))
    return std, up, low

std, up, low = err_ranges(x, y, popt)
```

```python
# Estimate lower and upper limits of the confidence range
import scipy.stats as stats
def err_ranges(x, y, popt):
    #using 95% confidence level
    alpha = 0.95

    residuals = y - y1

    z = stats.norm.ppf(1-(1-alpha)/2)

    lower = y - z * np.std(residuals)

    upper = y + z * np.std(residuals)
    return upper, lower

upper, lower = err_ranges(x, y, popt)
```
